```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from ipywidgets import interact, IntSlider

# Load the image
img_path = 'path/to/youre/image'  # Replace with your image path
img = Image.open(img_path)
img_gray = img.convert('L')
img_array = np.array(img_gray)

# Function to update the plot
def update_plot(x, y):
    if x < img_array.shape[1] and y < img_array.shape[0]:
        value = img_array[y, x]
        print(f"Pixel at ({x}, {y}): {value}")

        # Redraw the image
        fig, ax = plt.subplots()
        ax.imshow(img_gray, cmap='gray')
        ax.axis('off')

        # Draw a box around the selected pixel
        rect = plt.Rectangle((x-0.5, y-0.5), 1, 1, edgecolor='red', facecolor='none')
        ax.add_patch(rect)

        plt.show()
    else:
        print("Pixel coordinate out of range")

# Create sliders and update plot interactively
interact(update_plot, x=IntSlider(min=0, max=img_array.shape[1]-1, step=1, value=0),
         y=IntSlider(min=0, max=img_array.shape[0]-1, step=1, value=0))
```
